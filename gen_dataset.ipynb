```python
import matplotlib.pyplot as plt
from skimage import draw
import numpy as np
import random as rnd
import cv2
```

```python

arr = np.zeros((200, 200))
stroke = 3
# Create an outer and inner circle. Then subtract the inner from the outer.
radius = 80
inner_radius = radius - (stroke // 2) + (stroke % 2) - 1 
outer_radius = radius + ((stroke + 1) // 2)
ri, ci = draw.circle(150, 100, radius=inner_radius, shape=arr.shape)
ro, co = draw.circle(100, 100, radius=outer_radius, shape=arr.shape)
arr[ro, co] = 13
arr[ri, ci] = 150
plt.imshow(arr)
plt.show()
```

```python
img = np.zeros((256,1256,3), int)
for i in range(1000):
    r = rnd.randint(0,255)
    c = rnd.randint(0, 1256)
    r_rad = rnd.randint(10,20)
    c_rad = rnd.randint(10,20)
    val = rnd.randint(0,128)
    color = [ val for k in range(3)]
    img[draw.ellipse(r,c,r_rad,c_rad, shape=(256,1256))]=color
plt.imshow(img)
```

```python
img2 = np.array(img)
mask = np.zeros((256,1256), bool)
for i in range(20):
    r = rnd.randint(0,255)
    c = rnd.randint(0, 1256)
    r_rad = rnd.randint(30,50)
    c_rad = rnd.randint(30,50)
    #color = [rnd.randint(0,255) for k in range(3)]
    val = rnd.randint(128,256)
    color = [ val for k in range(3)]
    img2[draw.ellipse(r,c,r_rad,c_rad, shape=(256,1256))]=color
    mask[draw.ellipse(r,c,r_rad,c_rad, shape=(256,1256))]=1
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(25, 25))
axes[0].imshow(img2)
axes[1].imshow(mask)
plt.show()
```

```python

for k in range(5,6):
    kol = np.zeros((1000))
    img = np.zeros((256,1256,3), int)
    for i in range(1000):
        r = rnd.randint(0,255)
        c = rnd.randint(0, 1256)
        r_rad = rnd.randint(10,20)
        c_rad = rnd.randint(10,20)
        val = rnd.randint(0,128)
        color = [ val for k in range(3)]
        img[draw.ellipse(r,c,r_rad,c_rad, shape=(256,1256))]=color
    img2 = np.array(img)
    mask = np.zeros((256,1256), bool)
    for i in range(20):
        r = rnd.randint(0,255)
        c = rnd.randint(0, 1256)
        r_rad = rnd.randint(30,50)
        c_rad = rnd.randint(30,50)
        #color = [rnd.randint(0,255) for k in range(3)]
        val = rnd.randint(128,256)
        color = [ val for k in range(3)]
        for ii in range(1000):
            if not ((ii > c - c_rad and ii > c + c_rad) or (ii + 255 < c - c_rad and ii + 255 < c + c_rad)):
                kol[ii] += 1
        img2[draw.ellipse(r,c,r_rad,c_rad, shape=(256,1256))]=color
        mask[draw.ellipse(r,c,r_rad,c_rad, shape=(256,1256))]=1

    img_path = 'data/gen1/test/images/'
    mask_path = 'data/gen1/test/masks/'
    kol_path = 'data/gen1/kol/'
    for l in range (1000):
        ll = k * 1000 + l
        img_mask = np.zeros((256,256,3), int)
        img_mask[mask[:,l:l+256]>0]=[255,255,255]
        cv2.imwrite(img_path+str(ll)+'.jpg',img2[:,l:l+256])
        #print(img_path+str(l)+'.jpg')
        cv2.imwrite(mask_path+str(ll)+'.jpg',img_mask)
    np.save(kol_path+'5.npy', kol)
```
